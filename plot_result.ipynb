```python
import pandas as pd
# from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.dates as md
```

## 707 nm data

```python
df = pd.read_csv('data/relock_data_707.csv')
# df_ev = pd.read_csv('data/relock_data_event_707.csv')

df['time'] = pd.to_datetime(df['time'], unit='s', utc=True)
# df_ev['time'] = pd.to_datetime(df_ev['time'], unit='s', utc=True)

freq = df['freq'].to_numpy()

i = 0
while i < len(freq):

    if freq[i] > 0.01:
        if i > 0:
            freq[i] = freq[i-1]
        else:
            freq[i] = freq[i+1]
    
    # if df['time'][i] == df_ev['time'][0]:
    #     rel_i = i
    # if df['time'][i] == (df_ev['time'][0] + df_ev['long'][0]):
    #     rel_ii = i

    i += 1

freq *= 1e6
# rel_freq = freq[43:161]
# rel_t = df['time'].iloc[43:161]
# print(len(rel_freq))
```

```python
fig=plt.figure()
ax1=fig.add_subplot(111)
ax1.plot(df['time'], freq)
ax1.set_xlim([df['time'][0], df['time'][len(freq) - 1]])
ax1.set_ylim([-200, 200])
ax1.xaxis.set_major_locator(md.MinuteLocator())
ax1.xaxis.set_major_formatter(md.DateFormatter('%H:%M'))
plt.setp(ax1.xaxis.get_majorticklabels(), rotation = 90)
ax1.set_xlabel('Time')
ax1.set_ylabel('Frequency Difference [MHz]')
ax1.set_title('Relocking ECDL @ 423.91355 THz by Wavemeter')

# Create a Rectangle patch
rect = patches.Rectangle((df['time'][0], -50), (df['time'][len(freq) - 1] - df['time'][0]), 100, linewidth=1, edgecolor='lime', facecolor='lime')
# Add the patch to the Axes
ax1.add_patch(rect)
ax1.text(df['time'][5],-20,'Safe Range',color='b')

# ax2 = fig.add_axes([0.45, 0.57, 0.25, 0.25])
# ax2.plot(rel_t, rel_freq)
# ax2.set_ylim([0, 200])
# ax2.xaxis.set_major_locator(md.MinuteLocator())
# ax2.xaxis.set_major_formatter(md.DateFormatter('%S'))
# plt.setp(ax2.xaxis.get_majorticklabels(), rotation = 90)

plt.savefig('data/relock_data_707.png', bbox_inches='tight')
plt.show()
```
